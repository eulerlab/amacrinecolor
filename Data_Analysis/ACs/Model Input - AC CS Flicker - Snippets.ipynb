```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

Load data

```python
# Stimuli
uv_stim = np.load('Data/ACs/Stim/uv_flicker.npy')
green_stim = np.load('Data/ACs/Stim/green_flicker.npy')

# Responses 
flicker_responses = np.load('Data/ACs/Cluster Averages/flicker_avrg.npy')
```

Resample to 64 Hz

```python
# Stimulus
stim_duration = uv_stim.shape[1] * 0.0016 # In seconds
t_stim = np.arange(0, stim_duration, 0.0016) # Old frequency : 625
t_stim_new = np.arange(0, stim_duration, 1/64) # New frequency : 64
uv_stim_new = -99*np.ones((t_stim_new.shape[0],3))
green_stim_new = -99*np.ones((t_stim_new.shape[0],3))

for i in range(3):
    f = interpolate.interp1d(t_stim, uv_stim[i,:], bounds_error = False)
    uv_stim_new[:,i] = f(t_stim_new)
    
    f = interpolate.interp1d(t_stim, green_stim[i,:], bounds_error = False)
    green_stim_new[:,i] = f(t_stim_new)
```

Remove nans

```python
np.where(t_stim_new > t_stim[-1])
```

Make stimulus binary

```python
uv_stim_new_binary = uv_stim_new.copy()
uv_stim_new_binary[uv_stim_new<0.5] = 0
uv_stim_new_binary[uv_stim_new>=0.5] = 1

green_stim_new_binary = green_stim_new.copy()
green_stim_new_binary[green_stim_new<0.5] = 0
green_stim_new_binary[green_stim_new>=0.5] = 1
```

Plot

```python
# Plot stim
plt.figure(figsize=(19,3))
plt.subplot(1,3,1)
plt.plot(t_stim, uv_stim[0,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')

plt.subplot(1,3,2)
plt.plot(t_stim, uv_stim[1,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')

plt.subplot(1,3,3)
plt.plot(t_stim, uv_stim[2,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,2], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')
plt.show()

plt.figure(figsize=(19,3))
plt.subplot(1,3,1)
plt.plot(t_stim, green_stim[0,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')

plt.subplot(1,3,2)
plt.plot(t_stim, green_stim[1,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')

plt.subplot(1,3,3)
plt.plot(t_stim, green_stim[2,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,2], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')
plt.show()


plt.figure(figsize=(19,3))
plt.subplot(1,3,1)
plt.plot(t_stim, uv_stim[0,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')

plt.subplot(1,3,2)
plt.plot(t_stim, uv_stim[1,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')

plt.subplot(1,3,3)
plt.plot(t_stim, uv_stim[2,:], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,2], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')
plt.show()

plt.figure(figsize=(19,3))
plt.subplot(1,3,1)
plt.plot(t_stim, green_stim[0,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')

plt.subplot(1,3,2)
plt.plot(t_stim, green_stim[1,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')

plt.subplot(1,3,3)
plt.plot(t_stim, green_stim[2,:], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,2], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')
plt.show()

print('Making binary')
plt.figure(figsize=(11,3))
plt.subplot(1,2,1)
plt.plot(t_stim_new, uv_stim_new[:,0], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')

plt.subplot(1,2,2)
plt.plot(t_stim_new, uv_stim_new[:,1], '.', color = 'black')
plt.plot(t_stim_new, uv_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([30,35])
plt.xlabel('Time [s]')
plt.show()

plt.figure(figsize=(11,3))
plt.subplot(1,2,1)
plt.plot(t_stim_new, green_stim_new[:,0], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,0], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')

plt.subplot(1,2,2)
plt.plot(t_stim_new, green_stim_new[:,1], '.', color = 'black')
plt.plot(t_stim_new, green_stim_new_binary[:,1], '.', color = 'red')
plt.xlim([170,180])
plt.xlabel('Time [s]')
plt.show()
```

Resample to 64 Hz

```python
# Responses
assert stim_duration == flicker_responses.shape[1] * 0.0016

t_resp = np.arange(0, stim_duration, 0.0016) # Old frequency : 625
t_resp_new = np.arange(0, stim_duration, 1/64) # New frequency : 64

flicker_responses_new = -99*np.ones((flicker_responses.shape[0], t_resp_new.shape[0]))

for i in range(flicker_responses.shape[0]): # Iterate over number of clusters
    f = interpolate.interp1d(t_resp, flicker_responses[i,:], bounds_error = False)
    flicker_responses_new[i,:] = f(t_resp_new)
```

```python
np.where(t_stim_new > t_stim[-1])
```

Plot example window

```python
# Plot response
for i in range(flicker_responses.shape[0]):
    plt.figure(figsize=(11,3))
    plt.subplot(1,2,1)
    plt.plot(t_resp, flicker_responses[i,:], '.', color = 'black')
    plt.plot(t_resp_new, flicker_responses_new[i,:], '.', color = 'red')
    plt.xlim([30,35])
    plt.subplot(1,2,2)
    plt.plot(t_resp, flicker_responses[i,:], '.', color = 'black')
    plt.plot(t_resp_new, flicker_responses_new[i,:], '.', color = 'red')
    plt.xlim([170,180])
    plt.show()
```
